````python
from hypothesis import given , strategies as st
def count_sort1(nums):
  count = [0] * len(nums)
  nums_sorted = [0] * len(nums)
  for i in range(len(nums) - 1):
    for j in range(i + 1, len(nums)):
      if nums[i] > nums[j]:
        count[i] += 1
      elif nums[i] < nums[j]:
        count[j] += 1
  for i in range(len(nums)):
    nums_sorted[count[i]] = nums[i]
  return nums_sorted

def count_sort2(nums):
  count = [0] * len(nums)
  nums_sorted = [0] * len(nums)
  for i in range(len(nums) - 1):
    for j in range(i + 1, len(nums)):
      if nums[i] > nums[j]:
        count[i] += 1
      elif nums[i] <= nums[j]:
        count[j] += 1
  for i in range(len(nums)):
    nums_sorted[count[i]] = nums[i]
  return nums_sorted

@given(nums=st.lists(st.integers()))
#Unit Test
def test_Inversions(nums):
    count3=count_sort1(nums)
    count4=count_sort2(nums)
    assert count3==count4,f"nums= {nums}, count1 = {count3}, count2 ={count4}"

test_Inversions()
````

(a) Using the technique of divide-and-conquer, write a recursive program
to find the maximum value in a given (unsorted) list of numbers.

(b) Write the recurrence relation to find the time complexity of the algorithm.
Find a closed form expression for the time complexity.

````python
import sys
import random
def generate(l,n):
  while (len(l) < 20):
    x = random.randint(0, n)
    l.append(x)
  return l

def findMax(nums, left, right, max=-sys.maxsize):
 
    if left == right:               
        if max < nums[left]:           
            max = nums[left]
        return max
    if right - left == 1:           
        if nums[left] < nums[right]:      
            if max < nums[right]:      
                max = nums[right] 
        else:
            if max < nums[left]:       
                max = nums[left]
        return max
    mid = (left + right) // 2
    max = findMax(nums, left, mid, max)
    max = findMax(nums, mid + 1, right, max)
 
    return max
    
nums=[]
nums=generate(nums,100)
print(nums)
print("The Maximum number in given Array is : ",findMax(nums,0,len(nums)-1))
````
